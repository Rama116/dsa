```python
import numpy as np

# Sample data
data = [10, 20, 30, 40, 50]

# Statistical basics
mean = np.mean(data)
median = np.median(data)
std_dev = np.std(data)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard Deviation: {std_dev}")
```

```python
from sklearn.cluster import KMeans
import numpy as np

# Sample data
X = np.array([[1, 2], [1, 4], [1, 0],
              [4, 2], [4, 4], [4, 0]])

# K-Means clustering
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
print("Cluster Centers:", kmeans.cluster_centers_)
print("Labels:", kmeans.labels_)
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()
X = iris.data

# K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
print("Cluster Centers:", kmeans.cluster_centers_)
print("Labels:", kmeans.labels_)
```

```python
from pyclustering.cluster.kmedoids import kmedoids
import numpy as np

# Sample data
X = np.array([[1, 2], [1, 4], [1, 0],
              [4, 2], [4, 4], [4, 0]])

# Initial medoids
initial_medoids = [0, 3]

# K-Medoids clustering
kmedoids_instance = kmedoids(X, initial_medoids)
kmedoids_instance.process()
clusters = kmedoids_instance.get_clusters()

print("Clusters:", clusters)
print("Medoids:", kmedoids_instance.get_medoids())
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Sample data
data = [[1, 2], [3, 4], [1, 0], [4, 4]]

# Hierarchical clustering
linked = linkage(data, method='ward')
dendrogram(linked)
plt.title("Hierarchical Clustering Dendrogram")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

# Sample data
data = [[1, 2], [3, 4], [1, 0], [4, 4]]

# Agglomerative Clustering
agg_clustering = AgglomerativeClustering(n_clusters=2)
agg_clustering.fit(data)
print("Labels:", agg_clustering.labels_)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([1.5, 3.5, 5.5, 7.5, 9.5])

# Model
model = LinearRegression()
model.fit(X, y)

print("Slope:", model.coef_)
print("Intercept:", model.intercept_)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Data
X = np.array([[1, 1], [2, 2], [3, 3], [4, 4]])
y = np.array([2, 4, 6, 8])

# Model
model = LinearRegression()
model.fit(X, y)

print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np

# Data
X = np.array([[0], [1], [2], [3]])
y = np.array([0, 0, 1, 1])

# Model
model = LogisticRegression()
model.fit(X, y)

print("Prediction for [1.5]:", model.predict([[1.5]]))
```

```python
from sklearn.svm import SVC

# Data
X = [[0, 0], [1, 1]]
y = [0, 1]

# Model
clf = SVC(kernel='linear')
clf.fit(X, y)

print("Prediction:", clf.predict([[0.5, 0.5]]))
```

```python
from sklearn.svm import SVC

# Data
X = [[0, 0], [1, 1], [2, 0], [0, 2]]
y = [0, 0, 1, 1]

# Model
clf = SVC(kernel='rbf')
clf.fit(X, y)

print("Prediction:", clf.predict([[1, 1]]))
```

```python
from sklearn.naive_bayes import GaussianNB

# Data
X = [[1, 2], [3, 4], [5, 6]]
y = [0, 1, 1]

# Model
clf = GaussianNB()
clf.fit(X, y)

print("Prediction for [2, 3]:", clf.predict([[2, 3]]))
```

```python
from sklearn.tree import DecisionTreeClassifier

# Data
X = [[0, 0], [1, 1]]
y = [0, 1]

# Model
clf = DecisionTreeClassifier()
clf.fit(X, y)

print("Prediction:", clf.predict([[0.5, 0.5]]))
```

```python
from sklearn.decomposition import PCA
import numpy as np

# Data
X = np.array([[1, 2], [3, 4], [5, 6]])

# PCA
pca = PCA(n_components=1)
X_reduced = pca.fit_transform(X)
print("Reduced Data:", X_reduced)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# Data
X = [[1, 2], [3, 4], [5, 6]]
y = [0, 1, 1]

# LDA
lda = LinearDiscriminantAnalysis()
lda.fit(X, y)
print("Prediction for [2, 3]:", lda.predict([[2, 3]]))
```
